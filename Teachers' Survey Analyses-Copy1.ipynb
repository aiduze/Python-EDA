# EXPLORATORY DATA ANALYSIS ON TEACHER SURVEY 

## Background on the survey:

 ### A needs assessment survey was carried out to derive insights that can help design a tool that will improve teachers' output. The survey contained 11 mixed-type questions and 112 responses were gathered over the course of 3 days, and they have been presented for analysis.
 
### SURVEY QUESTIONS:

* What is your gender?
* What is your age range?
* What is your highest level of education?
* How many subjects do you teach in your grade?
* What grade do you teach?
* Which of these would you say are some of the main problems you face as a teacher (Choose all answers that apply)
* Do you use the continuous assessments scores only at the end of the session?
* Do you have a mobile phone
* If you have a mobile phone, what type is it?
* If you had a tool on my phone/tab that helps me teach better, what would you like it to help with (choose all the best options that apply to you)

### DATA DESCRIPTION:

The data was cleaned an prepped for analysis in excel by fixing data types and shortening the column names. The entries in the dataset are described as follows:

- **`Submission Date:`** Date that the survey response was submitted.

- **`Gender:`** Gender of the respondent.

- **`Age_Bracket:`** Age bracket of the respondent which is given as intervals 0-17, 18-24, 25-34, 35-44 or 45-54.
- **`Education:`** This is the highest level of education of the respondent. Options are: Lower than MSCE, MSCE, PTCE/Diploma, Bachelors Degree, Masters Degree.
- **`Number_of_subjects:`** Number of subjects taught by the teacher.
- **`Teaching_grade:`** Grade level of the children that the teacher teaches.
- **`Problems:`** Main problems being faced by the teacher.
- **`CA _use_end:`** Whether or not the teacher uses continuous assessment scores `only` at the end of the session. Yes/No
- **`Mobile_device:`** Whether or not the teacher own a mobile device. Yes/No
- **`Type:`** The type of mobile device and the options are Smartphone, keypad phone or no phone
- **`Preferred_area_of_help:`** Areas where the teacher feels a tool can help them improve their teaching.

```python
#Define the function for labelled barplots:
def labelled_barplot(data, feature, perc=False, n=None):
    """
    Barplot with percentage at the top

    data: dataframe
    feature: dataframe column
    perc: whether to display percentages instead of count (default is False)
    n: displays the top n category levels (default is None, i.e., display all levels)
    """

    total = len(data[feature])  # length of the column
    count = data[feature].nunique()
    if n is None:
        plt.figure(figsize=(count + 1, 5))
    else:
        plt.figure(figsize=(n + 1, 5))

    plt.xticks(rotation=90, fontsize=15)
    ax = sns.countplot(
        data=data,
        x=feature,
        palette="Paired",
        order=data[feature].value_counts().index[:n].sort_values(),
    )
    fig = ax.figure

    for p in ax.patches:
        if perc == True:
            label = "{:.1f}%".format(
                100 * p.get_height() / total
            )  # percentage of each class of the category
        else:
            label = p.get_height()  # count of each level of the category

        x = p.get_x() + p.get_width() / 2  # width of the plot
        y = p.get_height()  # height of the plot

        ax.annotate(
            label,
            (x, y),
            ha="center",
            va="center",
            size=12,
            xytext=(0, 5),
            textcoords="offset points",
        )  # annotate the percentage

    plt.show()  # show the plot
    
    return fig
```

```python
# Define the function for a stacked barplot: 
def stacked_barplot(data, predictor, target):
    """
    Print the category counts and plot a stacked bar chart

    data: dataframe
    predictor: independent variable
    target: target variable
    """
    count = data[predictor].nunique()
    sorter = data[target].value_counts().index[-1]
    tab1 = pd.crosstab(data[predictor], data[target], margins=True).sort_values(
        by=sorter, ascending=False
    )
    print(tab1)
    print("-" * 120)
    tab = pd.crosstab(data[predictor], data[target], normalize="index").sort_values(
        by=sorter, ascending=False
    )
    fig = tab.plot(kind="bar", stacked=True, figsize=(count + 5, 5))
    plt.legend(
        loc="lower left",
        frameon=False,
    )
    plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
    plt.show()

    return fig
```

## Looking closer at the columns with multiple selection entries:

### Analysing the 'Problems' column:

```python
Probls = df["Problems"].str.split("\n", expand=True) #Split the responses in the column and list them
```

```python
#Count the number of unique responses and display as a datafram:
df_prob = Probls.stack().value_counts().rename_axis('Unique_responses').reset_index(name ='Counts') #Stack and count the unique responses
df_prob
```

```python
#Create a dataframe of the distribution of responses from the options given
opt_prob = df_prob[(df_prob['Counts'] > 3)]
```

```python
#Create a bar chart to visualize:
plt.figure(figsize = (12,7))
sns.barplot(data=opt_prob, x='Unique_responses', y='Counts')
plt.xticks(rotation=90)
plt.show()
```

```python
#Create a dataframe of the distribution of 'Other' responses:
other_prob = df_prob[(df_prob['Counts'] <= 3)]
other_prob
```

### A wordcloud will help identify the problems that are most mentioned in 'Other' responses to the question.

```python
# Wordcloud
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator

# Create stopword list:
stopwords = set(STOPWORDS)
stopwords.update(["the","school","teacher","teachers","teaching","and", "no", "No", "much","problem","problems"]) #These words are bound to occur frequently and will not add value.

textt = " ".join(review for review in other_prob.Unique_responses)
wordcloud = WordCloud(stopwords=stopwords).generate(textt)

plt.figure(figsize = (12,9))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
#plt.savefig('wordcloud11.png')
plt.show()
```

* There are a lot of concerns about the learners.
* Security and promotions are the next most mentioned words.
