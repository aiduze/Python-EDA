```python
#Defining a labelled barplot:

def labeled_barplot(data, feature, perc=False, n=None, custom_order=None):
    """
    Barplot with percentage at the top

    data: dataframe
    feature: dataframe column
    perc: whether to display percentages instead of count (default is False)
    n: displays the top n category levels (default is None, i.e., display all levels)
    custom_order: custom order for the feature column (default is None)
    """

    total = len(data[feature])  # length of the column
    if custom_order:
        data[feature] = pd.Categorical(data[feature], categories=custom_order, ordered=True)
    if n is None:
        plt.figure(figsize=(len(data[feature].unique()) + 1, 5))
    else:
        plt.figure(figsize=(n + 1, 5))

    plt.xticks(rotation=90, fontsize=15)
    ax = sns.countplot(
        data=data,
        x=feature,
        palette="Paired",
        order=custom_order if custom_order else data[feature].value_counts().index[:n].sort_values(),
    )

    for p in ax.patches:
        if perc:
            label = "{:.1f}%".format(100 * p.get_height() / total)  # percentage of each class of the category
        else:
            label = p.get_height()  # count of each level of the category

        x = p.get_x() + p.get_width() / 2  # width of the plot
        y = p.get_height()  # height of the plot

        ax.annotate(label, (x, y), ha="center", va="center", size=12, xytext=(0, 5), textcoords="offset points")  # annotate the percentage

    plt.show()  # show the plot
```

### The following plots will analyse the respondents at Baseline against the profficiencies to visualize the outcomes:

```python
#Function defining a stacked barplot:
def stacked_barplot(data, predictor, target, custom_order=None):
    """
    Print the category counts and plot a stacked bar chart

    data: dataframe
    predictor: independent variable
    target: target variable
    custom_order: list specifying custom order of categories (optional)
    """
    count = data[predictor].nunique()
    sorter = data[target].value_counts().index[-1]
    tab1 = pd.crosstab(data[predictor], data[target], margins=True).sort_values(
        by=sorter, ascending=False
    )
    print(tab1)
    print("-" * 120)
    tab = pd.crosstab(data[predictor], data[target], normalize="index").sort_values(
        by=sorter, ascending=False
    )
    
    # If custom_order is provided, use it to order the categories
    if custom_order:
        tab = tab.loc[custom_order]
    
    fig = tab.plot(kind="bar", stacked=True, figsize=(count + 5, 5))
    plt.legend(
        loc="lower left",
        frameon=False,
    )
    plt.legend(loc="upper left", bbox_to_anchor=(1, 1))
    plt.show()

    return fig
```
